**Mean Square Error**

image.png

```python
import math
```

```python
def mse(actual,predict):
  error=0
  n=len(actual)
  for ac , pc in zip(actual,predict):
    error += (ac - pc) ** 2
  mean_square_error=error/n
  return mean_square_error
```

**Prediction 1**

```python
actual = [1, 0, 0, 1, 1]
predict = [0.5, 0.3, 0.9, 0.7, 0.2]
```

```python
result = mse(actual, predict)
print(result)
```

**Prediction 2**

```python
actual = [1, 1, 0, 1, 1]
predict = [0.8, 0.5, 0.3, 0.7, 0.2]
```

```python
result = mse(actual, predict)
print(result)
```

**Sparse Categorial CrossEntropy**

CE.JPG

```python
import math
```

```python
import numpy as np
```

```python
def sparse_categorical_cross_entropy_loss(y_actual, y_predict):
    loss = -np.sum(y_actual * np.log(y_predict))
    return loss
```

**Prediction 1**

```python
y_actual = [0 , 1 , 0]
y_predict = [0.2 , 0.45 , 0.35]
```

```python
loss = sparse_categorical_cross_entropy_loss(y_actual, y_predict)
print("Categorical Cross-Entropy Loss:", loss)
```

**Prediction 2**

```python
y_actual  = np.array([1, 0, 0])
y_predict = np.array([0.9, 0.05, 0.05])
```

```python
loss = sparse_categorical_cross_entropy_loss(y_actual, y_predict)
print("Categorical Cross-Entropy Loss:", loss)
```

**Binary Crossentropy**

image.png

```python
import math
import numpy as np
```

```python
def binary_cross_entropy_loss(y_actual, y_predict):
    n=len(y_actual)
    loss=-np.sum(y_actual*np.log(y_predict)+((1-y_actual)*(np.log(1-y_predict))))/n
    return loss
```

```python
y_actual=np.array([0,1,1,0])
y_predict=np.array([0.2,0.6,0.9,0.3])
```

```python
result = binary_cross_entropy_loss(y_actual, y_predict)
```
